```python
import sys
import pyspark
from pyspark.sql import SparkSession
from pyspark.sql import functions as F
from pyspark.sql import types as T

from pyspark.ml.feature import VectorAssembler, StandardScaler
from pyspark.ml.stat import Correlation

print("🔹 Cargando librerías...")
print(f"✅ PySpark version: {pyspark.__version__}")

spark = SparkSession.builder \
    .appName("Parcial_Fraude_Bancario") \
    .config("spark.driver.memory", "4g") \
    .getOrCreate()

print("✅ Sesión Spark creada correctamente")
```

```python
RUTA_DATOS = "../datos/"   # ajusta según tu estructura

ruta_train = RUTA_DATOS + "train.csv"
ruta_test = RUTA_DATOS + "test.csv"
ruta_sample = RUTA_DATOS + "sample_submission.csv"

print("📂 Rutas definidas:")
print(" - Train:", ruta_train)
print(" - Test:", ruta_test)
print(" - Sample submission:", ruta_sample)
```

```python
print("=" * 70)
print("📥 CARGA DE DATOS")
print("=" * 70)

train_df = spark.read.csv(ruta_train, header=True, inferSchema=True)
test_df = spark.read.csv(ruta_test, header=True, inferSchema=True)
sample_df = spark.read.csv(ruta_sample, header=True, inferSchema=True)

print("✅ Archivos cargados correctamente")
print("   - train.csv")
print("   - test.csv")
print("   - sample_submission.csv")
```

```python
print("=" * 70)
print("🔎 REVISIÓN INICIAL DE ESTRUCTURA")
print("=" * 70)

train_rows = train_df.count()
test_rows = test_df.count()
sample_rows = sample_df.count()

print("📌 TRAIN")
train_df.printSchema()
print(f"✅ Filas: {train_rows:,}")
print(f"✅ Columnas: {len(train_df.columns)}")

print("\n📌 TEST")
test_df.printSchema()
print(f"✅ Filas: {test_rows:,}")
print(f"✅ Columnas: {len(test_df.columns)}")

print("\n📌 SAMPLE SUBMISSION")
sample_df.printSchema()
print(f"✅ Filas: {sample_rows:,}")
print(f"✅ Columnas: {len(sample_df.columns)}")
```

```python
print("=" * 70)
print("🧩 VALIDACIÓN DE COLUMNAS ESPERADAS")
print("=" * 70)

expected_train_cols = ["Time"] + [f"V{i}" for i in range(1, 29)] + ["Amount", "Class"]
expected_test_cols = ["Time"] + [f"V{i}" for i in range(1, 29)] + ["Amount"]

train_ok = train_df.columns == expected_train_cols
test_ok = test_df.columns == expected_test_cols

print(f"✅ Train tiene columnas esperadas: {train_ok}")
print(f"✅ Test tiene columnas esperadas: {test_ok}")

if not train_ok:
    print("⚠️ Columnas actuales de TRAIN:")
    print(train_df.columns)

if not test_ok:
    print("⚠️ Columnas actuales de TEST:")
    print(test_df.columns)
```

```python
print("=" * 70)
print("🆔 CREACIÓN DE ID INTERNO")
print("=" * 70)

train_df = train_df.withColumn("id", F.monotonically_increasing_id())
test_df = test_df.withColumn("id", F.monotonically_increasing_id())

train_df = train_df.select(["id"] + [c for c in train_df.columns if c != "id"])
test_df = test_df.select(["id"] + [c for c in test_df.columns if c != "id"])

print("✅ Se creó la columna 'id' en train y test")
print("🔹 Ejemplo train:")
train_df.select("id", "Time", "Amount", "Class").show(5)
```

```python
print("=" * 70)
print("🧾 REVISIÓN DE TIPOS DE DATOS")
print("=" * 70)

tipos_df = spark.createDataFrame(
    [(c, str(train_df.schema[c].dataType)) for c in train_df.columns],
    ["columna", "tipo"]
)

print("✅ Tipos detectados en TRAIN:")
tipos_df.show(50, truncate=False)
```

```python
print("=" * 70)
print("🕳️ REVISIÓN DE VALORES NULOS")
print("=" * 70)

nulls_train = train_df.select([
    F.sum(F.col(c).isNull().cast("int")).alias(c) for c in train_df.columns
])

nulls_test = test_df.select([
    F.sum(F.col(c).isNull().cast("int")).alias(c) for c in test_df.columns
])

print("📌 Nulos en TRAIN")
nulls_train.show(truncate=False)

print("📌 Nulos en TEST")
nulls_test.show(truncate=False)

print("✅ Validación de nulos terminada")
```

```python
print("=" * 70)
print("📑 REVISIÓN DE DUPLICADOS")
print("=" * 70)

duplicados_train = train_df.count() - train_df.dropDuplicates().count()
duplicados_test = test_df.count() - test_df.dropDuplicates().count()

print(f"🔹 Duplicados en TRAIN: {duplicados_train}")
print(f"🔹 Duplicados en TEST: {duplicados_test}")

if duplicados_train == 0 and duplicados_test == 0:
    print("✅ No se detectaron duplicados exactos")
else:
    print("⚠️ Se detectaron duplicados; revisar antes de modelar")
```

```python
print("=" * 70)
print("🎯 DISTRIBUCIÓN DE LA VARIABLE OBJETIVO")
print("=" * 70)

total_train = train_df.count()

class_dist = train_df.groupBy("Class").count().orderBy("Class")
class_dist.show()

class_dist_pct = train_df.groupBy("Class").count() \
    .withColumn("porcentaje", F.round(F.col("count") / F.lit(total_train) * 100, 4)) \
    .orderBy("Class")

print("📌 Distribución porcentual")
class_dist_pct.show()

fraudes = train_df.filter(F.col("Class") == 1).count()
tasa_fraude = fraudes / total_train * 100

print(f"✅ Total observaciones train: {total_train:,}")
print(f"✅ Total fraudes: {fraudes:,}")
print(f"✅ Tasa de fraude: {tasa_fraude:.4f}%")
print("⚠️ Dataset altamente desbalanceado")
```

```python
print("=" * 70)
print("📊 RESUMEN DESCRIPTIVO BÁSICO")
print("=" * 70)

print("📌 TRAIN - Variables Time y Amount")
train_df.describe(["Time", "Amount"]).show()

print("📌 TEST - Variables Time y Amount")
test_df.describe(["Time", "Amount"]).show()
```

```python
print("=" * 70)
print("📏 CUANTILES DE VARIABLES CLAVE")
print("=" * 70)

quantiles_amount = train_df.approxQuantile("Amount", [0.0, 0.25, 0.5, 0.75, 0.9, 0.95, 0.99, 1.0], 0.001)
quantiles_time = train_df.approxQuantile("Time", [0.0, 0.25, 0.5, 0.75, 1.0], 0.001)

print("📌 Cuantiles de Amount:")
print(quantiles_amount)

print("\n📌 Cuantiles de Time:")
print(quantiles_time)

print("\n✅ Se observa la distribución general de monto y tiempo")
```

```python
print("=" * 70)
print("💰 ANÁLISIS DE AMOUNT POR CLASE")
print("=" * 70)

train_df.groupBy("Class").agg(
    F.count("*").alias("n"),
    F.round(F.mean("Amount"), 4).alias("mean_amount"),
    F.round(F.expr("percentile_approx(Amount, 0.5)"), 4).alias("median_amount"),
    F.round(F.min("Amount"), 4).alias("min_amount"),
    F.round(F.max("Amount"), 4).alias("max_amount")
).orderBy("Class").show(truncate=False)

print("✅ Comparación de montos entre transacciones normales y fraude")
```

```python
print("=" * 70)
print("⏱️ ANÁLISIS DE TIME POR CLASE")
print("=" * 70)

train_df.groupBy("Class").agg(
    F.count("*").alias("n"),
    F.round(F.mean("Time"), 4).alias("mean_time"),
    F.round(F.expr("percentile_approx(Time, 0.5)"), 4).alias("median_time"),
    F.round(F.min("Time"), 4).alias("min_time"),
    F.round(F.max("Time"), 4).alias("max_time")
).orderBy("Class").show(truncate=False)

print("✅ Comparación temporal entre clases realizada")
```

```python
print("=" * 70)
print("🧠 DEFINICIÓN DE FEATURES")
print("=" * 70)

feature_cols = ["Time"] + [f"V{i}" for i in range(1, 29)] + ["Amount"]

print(f"✅ Número de features numéricas: {len(feature_cols)}")
print("📌 Features incluidas:")
print(feature_cols)
```

```python
print("=" * 70)
print("🔧 TRANSFORMACIÓN DE AMOUNT")
print("=" * 70)

train_df = train_df.withColumn("Amount_log", F.log1p(F.col("Amount")))
test_df = test_df.withColumn("Amount_log", F.log1p(F.col("Amount")))

print("✅ Se creó la variable Amount_log con log1p(Amount)")
train_df.select("Amount", "Amount_log").show(5)
```

```python
print("=" * 70)
print("🔗 CÁLCULO DE CORRELACIONES")
print("=" * 70)

corr_cols = ["Time"] + [f"V{i}" for i in range(1, 29)] + ["Amount_log", "Class"]

assembler_corr = VectorAssembler(
    inputCols=corr_cols,
    outputCol="features_corr"
)

train_corr = assembler_corr.transform(train_df).select("features_corr")

corr_matrix = Correlation.corr(train_corr, "features_corr", "pearson").collect()[0][0]

print("✅ Matriz de correlación calculada correctamente en PySpark")
print(f"📌 Número de variables incluidas: {len(corr_cols)}")
```

```python
print("=" * 70)
print("🌡️ HEATMAP DE CORRELACIÓN")
print("=" * 70)

import pandas as pd
import matplotlib.pyplot as plt

corr_array = corr_matrix.toArray()
corr_df = pd.DataFrame(corr_array, index=corr_cols, columns=corr_cols)

plt.figure(figsize=(12, 10))
plt.imshow(corr_df, aspect="auto")
plt.colorbar(label="Correlación")

plt.xticks(range(len(corr_cols)), corr_cols, rotation=90)
plt.yticks(range(len(corr_cols)), corr_cols)

plt.title("Matriz de correlación")
plt.tight_layout()
plt.show()
```

### Interpretación general de la matriz de correlación

La matriz de correlación muestra ql, las variables tienen relaciones lineales bajas entreisí. Esto tiene sentido porque gran parte de las variables (`V1` a `V28`) provienen de una transformación PA, por lo que no se espera encontrar correlaciones muy altas entre ellas.

Tembién se observa que la variable objetivo `Class` no tiene una correlación extremadamente alta con una sola variable. Esto sugiere que el fraude no depende de una señal aisda , sino más bien de la combinación de varias variables al mismo t es decirbras, el problema parece requerir modelos que capten patrones conjuntos, y no solo relaciones lineales simples.

```python
print("=" * 70)
print("📊 TOP VARIABLES MÁS CORRELACIONADAS CON CLASS")
print("=" * 70)

import matplotlib.pyplot as plt

corr_class = corr_df["Class"].drop("Class").sort_values(key=abs, ascending=False)

print("✅ Variables más correlacionadas con Class:")
print(corr_class.head(10))

plt.figure(figsize=(10, 8))
corr_class.head(10).sort_values().plot(kind="barh")
plt.title("Top 10 variables más correlacionadas con Class")
plt.xlabel("Correlación de Pearson")
plt.ylabel("Variables")
plt.tight_layout()
plt.show()
```

### Interpretacion de las variables más correlacionadas con Class

Al revisar las variables con mayor correlación respecto a `Class`, se encuentra que algunas como `V17`, `V14`, `V12`, `V10`, `V16` y `V3` presentan las asociaciones más altas en valor absoluto. Aunque estas correlaciones no son muy grandes si nos indican que estas variables podrian aportar información util para distinguir entre transacciones normales y fraudulentas.

Además, el hecho de que varias de estas correlaciones sean negativas no es un problema, simplemente significa que a medida que aumenta el valor de esa variable, la probabilidad de fraude tiende a disminuir. En conjunto, este resultado sugiere que sí existen variables con señal estadística relevante, pero que el fraude sigue siendo un fenómeno complejo y difícil de separar con una sola feature.


```python
print("=" * 70)
print("📦 ENSAMBLADO DEL VECTOR DE FEATURES")
print("=" * 70)

model_features = ["Time"] + [f"V{i}" for i in range(1, 29)] + ["Amount_log"]

assembler = VectorAssembler(
    inputCols=model_features,
    outputCol="features_raw"
)

train_ml = assembler.transform(train_df)
test_ml = assembler.transform(test_df)

print("✅ Vector features_raw creado correctamente")
print(f"📌 Número de variables usadas en modelado: {len(model_features)}")
```

```python
print("=" * 70)
print("⚖️ ESCALAMIENTO DE VARIABLES")
print("=" * 70)

scaler = StandardScaler(
    inputCol="features_raw",
    outputCol="features",
    withMean=True,
    withStd=True
)

scaler_model = scaler.fit(train_ml)

train_ml = scaler_model.transform(train_ml)
test_ml = scaler_model.transform(test_ml)

print("✅ Features escaladas correctamente")
print("📌 Se usarán en los modelos que requieran variables en escala comparable")
```

```python
print("=" * 70)
print("✅ DATASETS FINALES PARA MODELADO")
print("=" * 70)

train_final = train_ml.select("id", "features", "Class")
test_final = test_ml.select("id", "features")

print("📌 TRAIN FINAL")
train_final.show(5, truncate=False)

print("📌 TEST FINAL")
test_final.show(5, truncate=False)

print("✅ Bases finales listas")
```

```python
print("=" * 70)
print("✂️ DIVISIÓN TRAIN / VALIDATION")
print("=" * 70)

SEED = 6913

print(f"📌 Semilla usada: {SEED}")

train_split, valid_split = train_final.randomSplit([0.8, 0.2], seed=SEED)

n_train_split = train_split.count()
n_valid_split = valid_split.count()

print(f"✅ Filas train interno: {n_train_split:,}")
print(f"✅ Filas validation: {n_valid_split:,}")
print(f"✅ Total verificado: {n_train_split + n_valid_split:,}")
```

```python
print("=" * 70)
print("🚨 DISTRIBUCIÓN DE FRAUDE EN TRAIN Y VALIDATION")
print("=" * 70)

print("📌 Train interno")
train_split.groupBy("Class").count().orderBy("Class").show()

print("📌 Validation")
valid_split.groupBy("Class").count().orderBy("Class").show()

fraudes_train = train_split.filter(F.col("Class") == 1).count()
fraudes_valid = valid_split.filter(F.col("Class") == 1).count()

print(f"✅ Fraudes en train interno: {fraudes_train:,}")
print(f"✅ Fraudes en validation: {fraudes_valid:,}")
```

```python
print("=" * 70)
print("⚖️ BALANCEO MEDIANTE PESOS AUTOMÁTICOS")
print("=" * 70)

conteos = train_split.groupBy("Class").count().collect()

n_total = sum(row["count"] for row in conteos)
n_clases = len(conteos)

pesos = {
    row["Class"]: n_total / (n_clases * row["count"])
    for row in conteos
}

print("📌 Pesos calculados por clase:")
print(pesos)

expr_peso = None
for clase, peso in pesos.items():
    condicion = (F.col("Class") == clase)
    if expr_peso is None:
        expr_peso = F.when(condicion, F.lit(float(peso)))
    else:
        expr_peso = expr_peso.when(condicion, F.lit(float(peso)))

train_split_w = train_split.withColumn("classWeightCol", expr_peso.otherwise(F.lit(1.0)))
valid_split_w = valid_split.withColumn("classWeightCol", expr_peso.otherwise(F.lit(1.0)))

print("✅ Columna classWeightCol creada en train y validation")
train_split_w.select("Class", "classWeightCol").show(10)
```

```python
print("=" * 70)
print("📋 RESUMEN FINAL DEL EDA Y PREPARACIÓN")
print("=" * 70)

fraudes = train_df.filter(F.col("Class") == 1).count()
tasa_fraude = fraudes / train_df.count() * 100

print(f"✅ Filas train: {train_df.count():,}")
print(f"✅ Filas test: {test_df.count():,}")
print(f"✅ Número de features para modelado: {len(model_features)}")
print(f"✅ Tasa de fraude en train: {tasa_fraude:.4f}%")
print(f"✅ Duplicados train: {duplicados_train}")
print("✅ Transformación aplicada: Amount_log")
print("✅ Escalamiento aplicado: StandardScaler")
print("✅ Split interno listo")
print("✅ Balanceo con pesos automáticos listo")
```

```python
print("=" * 70)
print("🤖 MODELO 1 - LOGISTIC REGRESSION (BASELINE)")
print("=" * 70)

from pyspark.ml.classification import LogisticRegression
from pyspark.ml.evaluation import BinaryClassificationEvaluator, MulticlassClassificationEvaluator
from pyspark.mllib.evaluation import MulticlassMetrics

print("✅ Librerías para Logistic Regression cargadas")
```

```python
print("=" * 70)
print("🏋️ ENTRENAMIENTO DEL MODELO BASE")
print("=" * 70)

print("📌 Configuración inicial del modelo:")
print(" - family = binomial")
print(" - regParam = 0.0")
print(" - elasticNetParam = 0.0")
print(" - maxIter = 100")
print(" - weightCol = classWeightCol")

lr = LogisticRegression(
    featuresCol="features",
    labelCol="Class",
    weightCol="classWeightCol",
    family="binomial",
    regParam=0.0,
    elasticNetParam=0.0,
    maxIter=100
)

lr_model = lr.fit(train_split_w)

print("✅ Logistic Regression entrenado correctamente")
```

```python
print("=" * 70)
print("🔮 PREDICCIONES EN VALIDATION")
print("=" * 70)

pred_lr = lr_model.transform(valid_split_w)

print("✅ Predicciones generadas correctamente")
print("📌 Columnas disponibles:")
print(pred_lr.columns)

pred_lr.select("id", "Class", "probability", "prediction").show(10, truncate=False)
```

```python
print("=" * 70)
print("📊 EVALUACIÓN DEL MODELO")
print("=" * 70)

eval_f1 = MulticlassClassificationEvaluator(
    labelCol="Class",
    predictionCol="prediction",
    metricName="f1"
)

eval_precision = MulticlassClassificationEvaluator(
    labelCol="Class",
    predictionCol="prediction",
    metricName="weightedPrecision"
)

eval_recall = MulticlassClassificationEvaluator(
    labelCol="Class",
    predictionCol="prediction",
    metricName="weightedRecall"
)

eval_auc_roc = BinaryClassificationEvaluator(
    labelCol="Class",
    rawPredictionCol="rawPrediction",
    metricName="areaUnderROC"
)

eval_auc_pr = BinaryClassificationEvaluator(
    labelCol="Class",
    rawPredictionCol="rawPrediction",
    metricName="areaUnderPR"
)

f1_lr = eval_f1.evaluate(pred_lr)
precision_lr = eval_precision.evaluate(pred_lr)
recall_lr = eval_recall.evaluate(pred_lr)
auc_roc_lr = eval_auc_roc.evaluate(pred_lr)
auc_pr_lr = eval_auc_pr.evaluate(pred_lr)

print("✅ Métricas calculadas")
print(f"📌 F1-score:  {f1_lr:.4f}")
print(f"📌 Precision: {precision_lr:.4f}")
print(f"📌 Recall:    {recall_lr:.4f}")
print(f"📌 AUC-ROC:   {auc_roc_lr:.4f}")
print(f"📌 AUC-PR:    {auc_pr_lr:.4f}")
```

```python
print("=" * 70)
print("🧾 MATRIZ DE CONFUSIÓN")
print("=" * 70)

confusion_lr = (
    pred_lr.groupBy("Class", "prediction")
    .count()
    .orderBy("Class", "prediction")
)

confusion_lr.show()

print("✅ Matriz de confusión generada")
```

```python
print("=" * 70)
print("🚨 MÉTRICAS ENFOCADAS EN LA CLASE POSITIVA (FRAUDE)")
print("=" * 70)

cm = pred_lr.select("Class", "prediction").rdd.map(lambda x: (float(x["prediction"]), float(x["Class"])))
metrics = MulticlassMetrics(cm)

precision_class1 = metrics.precision(1.0)
recall_class1 = metrics.recall(1.0)
f1_class1 = metrics.fMeasure(1.0)

print(f"📌 Precision clase 1: {precision_class1:.4f}")
print(f"📌 Recall clase 1:    {recall_class1:.4f}")
print(f"📌 F1 clase 1:        {f1_class1:.4f}")

print("\n✅ Estas métricas son especialmente importantes en detección de fraude")
```

```python
print("=" * 70)
print("📝 RESUMEN DEL MODELO BASE")
print("=" * 70)

print("✅ Logistic Regression se usó como modelo baseline")
print("✅ Se entrenó en PySpark con pesos automáticos por clase")
print("✅ Esto ayuda a manejar el fuerte desbalance del problema")
print(f"✅ F1 validation: {f1_lr:.4f}")
print(f"✅ AUC-ROC validation: {auc_roc_lr:.4f}")
print(f"✅ AUC-PR validation: {auc_pr_lr:.4f}")
print("📌 Este resultado servirá como referencia para comparar con los otros modelos")
```

### Interpretación del modelo Logistic Regression

En este primer modelo use una regresión logistica como un punto de partida para tener una referencia inicial del problema. el modelo mostró resultados bastante buenos en metricas globales, ya que obtuvo un F1 de 0.9860, una precisión de 0.9981, un recall de 0.9756, un AUC-ROC de 0.9890 y un AUC-PR de 0.7633.estas métricas hacen pensar que el modelo funciona muy bien pero en este caso hay que tener cuidado al interpretarlas, porque el dataset está muy desbalanceado y la mayoria de observaciones son transacciones normales.

Por eso importante revisar cómo se comporto el modelo especificamente con la clase de fraude. En ese punto se observa que el modelo alcanzo un recall de 0.9605 para la clase 1, lo cual es muy positivo, porque significa que logró detectar casi todos los fraudes reales. De hecho, en la matriz de confusión se ve que solo dejó escapar 3 fraudes y que identificó correctamente 73. Esto es valioso en un problema bancario porque dejar pasar un fraude suele ser más costoso que generar una alerta adicional.

Sin embargo, el modelo también tuvo una precisión de solo 0.0700 en la clase 1, lo que indica que muchas de las transacciones que marcó como fraude en realidad no lo eran. Esto se nota en los 970 falsos positivos, es decir, casos normales que el modelo clasificó como sospechosos.

```python
print("=" * 70)
print("🌲 MODELO 2 - RANDOM FOREST (BASELINE)")
print("=" * 70)

from pyspark.ml.classification import RandomForestClassifier
from pyspark.ml.evaluation import BinaryClassificationEvaluator, MulticlassClassificationEvaluator
from pyspark.mllib.evaluation import MulticlassMetrics

print("✅ Librerías para Random Forest cargadas")
```

```python
print("=" * 70)
print("🏋️ ENTRENAMIENTO DEL RANDOM FOREST BASE")
print("=" * 70)

print("📌 Configuración inicial del modelo:")
print(" - numTrees = 100")
print(" - maxDepth = 8")
print(" - minInstancesPerNode = 20")
print(" - seed = SEED")

rf = RandomForestClassifier(
    featuresCol="features",
    labelCol="Class",
    numTrees=100,
    maxDepth=8,
    minInstancesPerNode=20,
    seed=SEED
)

rf_model = rf.fit(train_split)

print("✅ Random Forest entrenado correctamente")
```

```python
print("=" * 70)
print("🔮 PREDICCIONES EN VALIDATION")
print("=" * 70)

pred_rf = rf_model.transform(valid_split)

print("✅ Predicciones generadas correctamente")
print("📌 Columnas disponibles:")
print(pred_rf.columns)

pred_rf.select("id", "Class", "probability", "prediction").show(10, truncate=False)
```

```python
print("=" * 70)
print("🧠 MODELO 3 - STACKING MANUAL")
print("=" * 70)
print("📌 Modelos base utilizados:")
print(" - Logistic Regression")
print(" - Random Forest")
print("📌 Estrategia:")
print(" - Combinar manualmente las probabilidades de ambos modelos")
print(" - Evaluar la predicción final sobre validation")
print("✅ Inicio del stacking manual")
```

```python
print("=" * 70)
print("📤 EXTRACCIÓN DE PROBABILIDADES - LOGISTIC REGRESSION")
print("=" * 70)

from pyspark.ml.functions import vector_to_array

pred_lr_stack = pred_lr.withColumn(
    "prob_lr",
    vector_to_array("probability")[1]
).select("id", "Class", "prob_lr")

pred_lr_stack.show(5, truncate=False)

print("✅ Probabilidades de Logistic Regression listas")
```

```python
print("=" * 70)
print("📤 EXTRACCIÓN DE PROBABILIDADES - RANDOM FOREST")
print("=" * 70)

from pyspark.ml.functions import vector_to_array

pred_rf_stack = pred_rf.withColumn(
    "prob_rf",
    vector_to_array("probability")[1]
).select("id", "prob_rf")

pred_rf_stack.show(5, truncate=False)

print("✅ Probabilidades de Random Forest listas")
print("=" * 70)
print("🔗 UNIÓN DE PREDICCIONES BASE")
print("=" * 70)

stack_df = pred_lr_stack.join(pred_rf_stack, on="id", how="inner")

print(f"✅ Filas combinadas: {stack_df.count():,}")
stack_df.show(5, truncate=False)
```

```python
print("=" * 70)
print("⚖️ COMBINACIÓN MANUAL DE PROBABILIDADES")
print("=" * 70)

peso_lr = 0.5
peso_rf = 0.5

print(f"📌 Peso Logistic Regression: {peso_lr}")
print(f"📌 Peso Random Forest: {peso_rf}")

stack_df = stack_df.withColumn(
    "prob_stack",
    (F.col("prob_lr") * F.lit(peso_lr)) + (F.col("prob_rf") * F.lit(peso_rf))
)

stack_df.select("id", "Class", "prob_lr", "prob_rf", "prob_stack").show(10, truncate=False)

print("✅ Probabilidad final del stacking creada")
```

```python
print("=" * 70)
print("🎯 GENERACIÓN DE PREDICCIÓN FINAL")
print("=" * 70)

umbral_stack = 0.5
print(f"📌 Umbral inicial: {umbral_stack}")

stack_df = stack_df.withColumn(
    "prediction",
    F.when(F.col("prob_stack") >= F.lit(umbral_stack), F.lit(1.0)).otherwise(F.lit(0.0))
)

stack_df.select("id", "Class", "prob_stack", "prediction").show(10, truncate=False)

print("✅ Predicción final generada")
```

```python
print("=" * 70)
print("🧱 PREPARACIÓN DEL DATAFRAME PARA EVALUACIÓN")
print("=" * 70)

stack_eval = stack_df.withColumn("rawPrediction", F.col("prob_stack")) \
    .withColumn("label", F.col("Class"))

stack_eval.select("id", "Class", "prob_stack", "prediction", "rawPrediction").show(5, truncate=False)

print("✅ Dataset de evaluación listo")
```

```python
print("=" * 70)
print("📊 EVALUACIÓN DEL STACKING MANUAL")
print("=" * 70)

eval_f1 = MulticlassClassificationEvaluator(
    labelCol="Class",
    predictionCol="prediction",
    metricName="f1"
)

eval_precision = MulticlassClassificationEvaluator(
    labelCol="Class",
    predictionCol="prediction",
    metricName="weightedPrecision"
)

eval_recall = MulticlassClassificationEvaluator(
    labelCol="Class",
    predictionCol="prediction",
    metricName="weightedRecall"
)

eval_auc_roc = BinaryClassificationEvaluator(
    labelCol="Class",
    rawPredictionCol="rawPrediction",
    metricName="areaUnderROC"
)

eval_auc_pr = BinaryClassificationEvaluator(
    labelCol="Class",
    rawPredictionCol="rawPrediction",
    metricName="areaUnderPR"
)

f1_stack = eval_f1.evaluate(stack_eval)
precision_stack = eval_precision.evaluate(stack_eval)
recall_stack = eval_recall.evaluate(stack_eval)
auc_roc_stack = eval_auc_roc.evaluate(stack_eval)
auc_pr_stack = eval_auc_pr.evaluate(stack_eval)

print("✅ Métricas calculadas")
print(f"📌 F1-score:  {f1_stack:.4f}")
print(f"📌 Precision: {precision_stack:.4f}")
print(f"📌 Recall:    {recall_stack:.4f}")
print(f"📌 AUC-ROC:   {auc_roc_stack:.4f}")
print(f"📌 AUC-PR:    {auc_pr_stack:.4f}")
```

```python
print("=" * 70)
print("🧾 MATRIZ DE CONFUSIÓN")
print("=" * 70)

confusion_stack = (
    stack_eval.groupBy("Class", "prediction")
    .count()
    .orderBy("Class", "prediction")
)

confusion_stack.show()

print("✅ Matriz de confusión generada")
```

```python
print("=" * 70)
print("🚨 MÉTRICAS ENFOCADAS EN LA CLASE POSITIVA (FRAUDE)")
print("=" * 70)

cm_stack = stack_eval.select("Class", "prediction").rdd.map(lambda x: (float(x["prediction"]), float(x["Class"])))
metrics_stack = MulticlassMetrics(cm_stack)

precision_class1_stack = metrics_stack.precision(1.0)
recall_class1_stack = metrics_stack.recall(1.0)
f1_class1_stack = metrics_stack.fMeasure(1.0)

print(f"📌 Precision clase 1: {precision_class1_stack:.4f}")
print(f"📌 Recall clase 1:    {recall_class1_stack:.4f}")
print(f"📌 F1 clase 1:        {f1_class1_stack:.4f}")

print("✅ Evaluación específica sobre fraude completada")
```

### Interpretación del Stacking Manual

El modelo de stacking manual combinó las probabilidades de la regresión logística y del Random Forest, con la idea de aprovechar lo mejor de los dos modelos. En los resultados se vio que este enfoque tuvo un desempeño bastante bueno, sobre todo porque logró aumentar la capacidad para detectar fraudes si lo comparamos con el Random Forest por sí solo. Eso se nota en el recall de 0.8816 para la clase 1, lo que quiere decir que pudo identificar una mayor parte de las transacciones fraudulentas.

Sin embargo, esa mejora en recall también vino con una pequeña caída en la precisión. La precisión de la clase 1 fue 0.7283, entonces aunque el modelo detectó más fraudes, también produjo más falsas alarmas que el Random Forest. En términos más generales, el F1 de la clase 1 fue 0.7976, que quedó un poco por debajo del 0.8201 que obtuvo el Random Forest. Entonces, aunque el stacking mostró un resultado sólido y claramente mejoró frente a la regresión logística, al final no alcanzó a superar al Random Forest como el mejor modelo.

```python
print("=" * 80)
print("🧪 FUNCIÓN PARA EXPERIMENTOS DE STACKING MANUAL")
print("=" * 80)

from pyspark.mllib.evaluation import MulticlassMetrics
from pyspark.ml.evaluation import BinaryClassificationEvaluator, MulticlassClassificationEvaluator

def evaluar_stacking(pred_lr, pred_rf, peso_lr=0.5, peso_rf=0.5, umbral=0.5):
    from pyspark.ml.functions import vector_to_array
    from pyspark.sql import functions as F
    
    # Extraer probabilidades
    pred_lr_stack = pred_lr.withColumn(
        "prob_lr",
        vector_to_array("probability")[1]
    ).select("id", "Class", "prob_lr")
    
    pred_rf_stack = pred_rf.withColumn(
        "prob_rf",
        vector_to_array("probability")[1]
    ).select("id", "prob_rf")
    
    # Unir
    stack_df = pred_lr_stack.join(pred_rf_stack, on="id", how="inner")
    
    # Combinar probabilidades
    stack_df = stack_df.withColumn(
        "prob_stack",
        (F.col("prob_lr") * F.lit(peso_lr)) + (F.col("prob_rf") * F.lit(peso_rf))
    )
    
    # Predicción final
    stack_df = stack_df.withColumn(
        "prediction",
        F.when(F.col("prob_stack") >= F.lit(umbral), F.lit(1.0)).otherwise(F.lit(0.0))
    )
    
    # Preparar para evaluación
    stack_eval = stack_df.withColumn("rawPrediction", F.col("prob_stack"))
    
    # Evaluadores
    eval_f1 = MulticlassClassificationEvaluator(
        labelCol="Class",
        predictionCol="prediction",
        metricName="f1"
    )
    
    eval_precision = MulticlassClassificationEvaluator(
        labelCol="Class",
        predictionCol="prediction",
        metricName="weightedPrecision"
    )
    
    eval_recall = MulticlassClassificationEvaluator(
        labelCol="Class",
        predictionCol="prediction",
        metricName="weightedRecall"
    )
    
    eval_auc_roc = BinaryClassificationEvaluator(
        labelCol="Class",
        rawPredictionCol="rawPrediction",
        metricName="areaUnderROC"
    )
    
    eval_auc_pr = BinaryClassificationEvaluator(
        labelCol="Class",
        rawPredictionCol="rawPrediction",
        metricName="areaUnderPR"
    )
    
    f1_general = eval_f1.evaluate(stack_eval)
    precision_general = eval_precision.evaluate(stack_eval)
    recall_general = eval_recall.evaluate(stack_eval)
    auc_roc = eval_auc_roc.evaluate(stack_eval)
    auc_pr = eval_auc_pr.evaluate(stack_eval)
    
    # Métricas clase 1
    cm_stack = stack_eval.select("Class", "prediction").rdd.map(
        lambda x: (float(x["prediction"]), float(x["Class"]))
    )
    metrics_stack = MulticlassMetrics(cm_stack)
    
    precision_class1 = metrics_stack.precision(1.0)
    recall_class1 = metrics_stack.recall(1.0)
    f1_class1 = metrics_stack.fMeasure(1.0)
    
    return {
        "peso_lr": peso_lr,
        "peso_rf": peso_rf,
        "umbral": umbral,
        "F1_general": f1_general,
        "Precision_general": precision_general,
        "Recall_general": recall_general,
        "AUC_ROC": auc_roc,
        "AUC_PR": auc_pr,
        "Precision_clase1": precision_class1,
        "Recall_clase1": recall_class1,
        "F1_clase1": f1_class1
    }

print("✅ Función creada correctamente")
```

```python
print("=" * 80)
print("🚀 EJECUCIÓN DE EXPERIMENTOS")
print("=" * 80)

experimentos = []

# Experimento 1
exp1 = evaluar_stacking(pred_lr, pred_rf, peso_lr=0.5, peso_rf=0.5, umbral=0.5)
experimentos.append(exp1)
print("✅ Experimento 1 listo")

# Experimento 2
exp2 = evaluar_stacking(pred_lr, pred_rf, peso_lr=0.3, peso_rf=0.7, umbral=0.5)
experimentos.append(exp2)
print("✅ Experimento 2 listo")

# Experimento 3
exp3 = evaluar_stacking(pred_lr, pred_rf, peso_lr=0.2, peso_rf=0.8, umbral=0.5)
experimentos.append(exp3)
print("✅ Experimento 3 listo")

# Experimento 4
exp4 = evaluar_stacking(pred_lr, pred_rf, peso_lr=0.4, peso_rf=0.6, umbral=0.4)
experimentos.append(exp4)
print("✅ Experimento 4 listo")

# Experimento 5
exp5 = evaluar_stacking(pred_lr, pred_rf, peso_lr=0.4, peso_rf=0.6, umbral=0.6)
experimentos.append(exp5)
print("✅ Experimento 5 listo")

print("\n🎉 Todos los experimentos terminaron")
```

```python
print("=" * 80)
print("📝 BITÁCORA DE EXPERIMENTOS ")
print("=" * 80)

bitacora_parcial = pd.DataFrame({
    "#": [1, 2, 3, 4, 5],
    "Hiperparámetro modificado": [
        "Peso Logistic",
        "Peso Logistic",
        "Peso Logistic",
        "Umbral",
        "Umbral"
    ],
    "Valor anterior": [
        "0.5",
        "0.3",
        "0.2",
        "0.5",
        "0.4"
    ],
    "Valor nuevo": [
        "0.3 (RF=0.7)",
        "0.2 (RF=0.8)",
        "0.4 (RF=0.6)",
        "0.4",
        "0.6"
    ],
    "F1 (validation)": [
        bitacora_df.loc[0, "F1_clase1"],
        bitacora_df.loc[1, "F1_clase1"],
        bitacora_df.loc[2, "F1_clase1"],
        bitacora_df.loc[3, "F1_clase1"],
        bitacora_df.loc[4, "F1_clase1"]
    ],
    "Decisión": [
        "Baseline del stacking",
        "Comparar más peso a RF",
        "Probar dominancia de RF",
        "Aumenta sensibilidad",
        "Aumenta precisión"
    ]
})

bitacora_parcial["F1 (validation)"] = bitacora_parcial["F1 (validation)"].round(4)

display(bitacora_parcial)
```

En la bitácora se observa que el stacking mejora cuando se reduce el peso de la regresión logística y se le da mayor participación al Random Forest. Esto sugiere que el Random Forest aporta una señal más sólida para distinguir fraude. También se ve que bajar demasiado el umbral no ayudó, porque aumentó la sensibilidad pero redujo el balance general. En cambio, mantener un umbral de 0.5 o incluso subirlo a 0.6 permitió obtener un mejor F1 en la clase positiva.
